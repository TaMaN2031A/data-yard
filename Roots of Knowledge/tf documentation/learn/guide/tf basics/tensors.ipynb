```python
import tensorflow as tf
import numpy as np
```

```python
#  single value, and no "axes"."scalar" or "rank-0"
rank_0_tensor = tf.constant(4)
print(rank_0_tensor)
```

```python
rank_1_tensor = tf.constant([2, 3, 4])
print(rank_1_tensor)
```

```python
rank_2_tensor = tf.constant([
    [1, 2],
    [3, 4],
    [5, 6]
], dtype=tf.int64)
print(rank_2_tensor)
```
