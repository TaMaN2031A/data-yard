```python
a = tf.constant([[1, 2],
                 [3, 4]])
b = tf.constant([[1, 1],
                 [1, 1]])

print(tf.add(a, b))
print(tf.multiply(a, b))
print(tf.matmul(a, b))
```

```python
print(a + b)
print(a * b)
print(a @ b)
```
