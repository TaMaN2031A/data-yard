# Naive Bayes for Sentiment Analysis

```python
import re
import math
import nltk

import pandas as pd
import numpy as np
from numpy.lib.function_base import vectorize
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
from collections import defaultdict
```

```python
data = pd.read_csv('../../datasets/IMDB Dataset.csv')
data
```

```python
def remove_tags(string):
    result = re.sub(r'<[^>]+>', '', string) # Remove HTML tags
    result = re.sub('https?://\S+', '', result) # Remove URLs
    result = re.sub(r'[^\w\s'']', ' ', result) # remove nonalphanumeric chars
    result = result.lower()
    return result
```

```python
data['review'] = data['review'].apply(lambda cw: remove_tags(cw))
```

```python
# nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words
```

```python
stop_words = set(stop_words)
```

```python
data['review'][1]
```

```python
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return " ".join([
        lemmatizer.lemmatize(x)
        for x in w_tokenizer.tokenize(text)
    ])

data['review'] = data['review'].apply(lemmatize_text)
```

```python
data['review'] = data['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
data['review'][5]
```

```python
reviews = data['review'].values
labels = data['sentiment'].values
```

```python
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)
```

```python
train_sentences, test_sentences, train_labels, test_labels = train_test_split(reviews, encoded_labels, test_size=0.2, random_state=42, stratify=encoded_labels)
```

```python
len({x for w in reviews for x in w.split()})
```

```python
vectorizer = CountVectorizer(max_features=1000, min_df=5, stop_words='english', max_df=0.8)

x = vectorizer.fit_transform(train_sentences)
vocab = vectorizer.get_feature_names_out()
x = x.toarray()
```

```python
word_count = {l: defaultdict(int) for l in range(2)}

for i in range(x.shape[0]):
    l = train_labels[i]
    for j in range(len(vocab)):
        count = x[i, j]
        if count > 0:
            word_count[l][vocab[j]] += count
```

```python
def laplace_smoothing(n_label_items, vocab, word_count, word, text_label):
    a = word_count[text_label][word] + 1
    b = n_label_items[text_label] + len(vocab)
    return math.log(a / b)
```

```python
def group_by_label(x, y, labels):
    data = {}
    for l in labels:
        data[l] = x[np.where(y == l)]
    return data
```

```python
def fit(x, y, labels):
    n_label_items = {}
    log_label_priors = {}
    n = len(x)
    grouped_data = group_by_label(x, y, labels)
    for label, data in grouped_data.items():
        n_label_items[label] = len(data)
        log_label_priors[label] = math.log(n_label_items[label] / n)

    return n_label_items, log_label_priors
```

```python
def predict(n_label_items, vocab, word_counts, log_label_priors, labels, x):
    result = []
    for text in x:
        label_scores = {l: log_label_priors[l] for l in labels}
        words = set(w_tokenizer.tokenize(text))
        for word in words:
            if word not in vocab: continue
            for l in labels:
                log_w_given_l = laplace_smoothing(n_label_items, vocab, word_count, word, l)
                label_scores[l] += log_w_given_l
        result.append(max(label_scores, key=label_scores.get))

    return result
```

```python
labels = [0,1]
n_label_items, log_label_priors = fit(train_sentences,train_labels,labels)
pred = predict(n_label_items, vocab, word_count, log_label_priors, labels, test_sentences)
```

```python
print("Accuracy of prediction on test set : ", accuracy_score(test_labels,pred))
```
