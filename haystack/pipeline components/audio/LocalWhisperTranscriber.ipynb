```python
import requests
from haystack.components.audio import LocalWhisperTranscriber
```

```python
respose = requests.get("https://ia903102.us.archive.org/19/items/100-Best--Speeches/EK_19690725_64kb.mp3")
with open("podcast.mp3", "wb") as f:
    f.write(respose.content)
```

```python
transcriber = LocalWhisperTranscriber(model="tiny")
transcriber.warm_up()
```

```python
transcription = transcriber.run(sources=["./podcast.mp3"])
print(transcription["documents"][0].content)
```

```python
from haystack.components.audio import LocalWhisperTranscriber
from haystack.components.fetchers import LinkContentFetcher
from haystack import Pipeline

pip = Pipeline()
pip.add_component('fetcher', LinkContentFetcher())
pip.add_component('transcriber', LocalWhisperTranscriber(model="tiny"))

pip.connect('fetcher', 'transcriber')

result = pip.run(
        data={"fetcher": {"urls": ["https://ia903102.us.archive.org/19/items/100-Best--Speeches/EK_19690725_64kb.mp3"]}}
)

print(result['transcriber']['documents'][0].content)
```
