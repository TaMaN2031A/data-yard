```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
iris = iris.frame
```

```python
iris.head()
```

```python
iris.assign(sepal_ratio = iris['sepal length (cm)'] / iris['sepal width (cm)'])
```

```python
iris.assign(sepal_ratio = lambda x: x['sepal length (cm)'] / x['sepal width (cm)'])
```

```python
(
    iris.query(" `sepal length (cm)` > 5")
    .assign(
        sepal_ratio = lambda x: x['sepal width (cm)'] / x['sepal length (cm)'],
        petal_ratio = lambda x: x['petal width (cm)'] / x['petal length (cm)'],
    )
    .plot(kind='scatter', x='sepal_ratio', y='petal_ratio')
)
```
