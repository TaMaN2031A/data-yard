Automatic differentiation

```python
x = tf.Variable(1.0)

def f(x):
    y = x**2 + x*2 - 5
    return y
```

```python
f(x)
```

```python
with tf.GradientTape() as tape:
    y = f(x)

g_x = tape.gradient(y, x)
g_x
```
